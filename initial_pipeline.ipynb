CONVERSION TO MARKDOWN FORMAT

```python
def pdf_to_markdown(pdf_path, md_path):
    doc = fitz.open(pdf_path)
    lines = []

    # umbral simple de heading por tamaГ±o de fuente
    H1 = 18
    H2 = 14

    for page in doc:
        data = page.get_text("dict")
        for block in data.get("blocks", []):
            if block.get("type", 0) != 0:
                continue  # ignora imГЎgenes por ahora

            for line in block.get("lines", []):
                spans = line.get("spans", [])
                if not spans:
                    continue

                text = "".join(s["text"] for s in spans).strip()
                if not text:
                    continue

                max_size = max(s["size"] for s in spans)

                if max_size >= H1:
                    lines.append(f"# {text}")
                elif max_size >= H2:
                    lines.append(f"## {text}")
                else:
                    # lista simple
                    if text.startswith(("вЂў", "-", "*")):
                        lines.append(f"- {text.lstrip('вЂў-* ').strip()}")
                    else:
                        lines.append(text)

            lines.append("")  # separaciГіn de bloque

    with open(md_path, "w", encoding="utf-8") as f:
        f.write("\n".join(lines))

if __name__ == "__main__":
    pdf_to_markdown("my_rag_test/input/1801.08322v4.pdf", "08322v4.md")
    pdf_to_markdown("my_rag_test/input/2408.12966v1.pdf", "12966v1.md")
    
```

```python
pdf_path = "2408.12966v1.pdf"
txt_path = "12966v1.txt"

with fitz.open(pdf_path) as doc, open(txt_path, "w", encoding="utf-8") as out:
    for pagina in doc:
        texto = pagina.get_text("text")
        out.write(texto)
        out.write("\n\n")

print(f"Listo: {txt_path}")
```
